```python
from urllib.request import urlopen
```

```python
story = urlopen('https://sherlock-holm.es/stories/plain-text/scan.txt')
```

```python
story_words = []
for line in story:
    line_words = line.split()
    for word in line_words:
        story_words.append(word)
```

```python
story.close()
```

```python
print(story_words)
```

```python
story = urlopen('https://sherlock-holm.es/stories/plain-text/scan.txt')
```

```python
story_words = []
for line in story:
    line_words = line.decode('utf8').split()
    for word in line_words:
        story_words.append(word)
```

```python
story.close()
```

```python
print(story_words)
```
